```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

Fruits = {
    "Fruit" : ["Apple","Banana","Mango","Orange","Grapes"],
    "Price" : [1500,600,4000,2000,700],
    "Calories" :[13000,50000,20070,6000,40020],
    "Weight" : [10,50,20,60,80],
    "NutritionalRating" : [3,3.5,4.6,2,4],
    "Protein" : [1.2,0.4,4,1,0.9]
}

df = pd.DataFrame(Fruits)
print(display(df))

# Line Chart
plt.plot(df['NutritionalRating'],df['Calories'],marker='s')
plt.title("Nutritional Rating Vs Calories")
plt.xlabel("Nutritional Rating")
plt.ylabel("Calories")
plt.show()

plt.figure(figsize=(16,9))
plt.plot(df['NutritionalRating'],df['Price'],marker='s',color="red",label="NutritionalRating")
plt.plot(df['Protein'],df['Price'],marker='o',color="blue",label="Protein")
plt.title("Nutritional Rating Vs Price")
plt.xlabel("Nutritional Rating")
plt.ylabel("Price")
plt.legend()
plt.show()

# Bar Chart
plt.bar(df['Fruit'],df['Calories'],color='yellow')
plt.title("Calories Vs Fruit")
plt.ylabel("Calories")
plt.xlabel("Fruit")
plt.show()

# Multibar graph
x = np.arange(len(df['Fruit']))
width = 0.35
plt.bar(x- width/2,df['Price'],width,label='Price Of The Fruit',edgecolor='black')
plt.bar(x+width/2,df['Calories'],width,label='Calories Of The Fruit',edgecolor='black')
plt.title("Price Vs Calories")
plt.xlabel("Fruit")
plt.ylabel("Values")

plt.xticks(x,df['Fruit'])
plt.title("Price VS Calories")
plt.xlabel("Fruit")
plt.ylabel("Values")
plt.legend()
plt.grid()
plt.show()

# Advance Bar Graph
colors = ['red','yellow','gold','orange','lightgreen']
bars = plt.bar(df['Fruit'],df['Calories'],color=colors,edgecolor='black')
for bar in bars:
  y=bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2,y,str(y),ha='center',va='bottom')
plt.title("Fruit Vs Calories")
plt.ylabel("Calories")
plt.xlabel("Fruit")
plt.grid(axis='y')
plt.show()

# Histogram
data1 = {
    "Weight" : [10,50,20,60,80,98,767,57,878,467,76,896,34,454,76]
    }
df1 = pd.DataFrame(data1)
plt.hist(df1['Weight'],bins=30,color='skyblue',edgecolor="black",alpha=0.75)
plt.title("Histogram")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.show()

# Advanced Histograms
data1 = {
    "Weight" : [10,50,20,60,80,98,767,57,878,467,76,896,34,454,76]
    }
df1 = pd.DataFrame(data1)
plt.hist(df1['Weight'],bins=30,color='yellow',edgecolor="black")
plt.title("Histogram")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.grid()
plt.show()

df['Taste']=['Sweet','Sweet','Sour','Sweet','Sour']
colors = ['green' if taste == 'Sweet' else 'red' for taste in df['Taste']]
plt.scatter(df['Fruit'],df["Price"],c = colors,marker='s')
plt.show()

# Graphs using Seaborn
data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'Mon', 'Tue', 'Wed'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': [200, 220, 250, 270, 300, 350, 400, 280, 260, 290],
    'Profit': [20, 25, 30, 35, 50, 60, 80, 40, 38, 45],
    'Customers': [30, 35, 40, 45, 50, 65, 80, 55, 48, 52],
    'Category': ['Food', 'Clothing', 'Electronics', 'Food', 'Clothing', 'Electronics', 'Food', 'Clothing', 'Electronics', 'Food']
}

df = pd.DataFrame(data)
print(df)

sns.lineplot(x='Day',y='Sales',data=df)
plt.title("Seaborn Line Plot")
plt.show()

sns.histplot(df['Sales'],bins=7,color='yellow')
plt.title("Seaborn Hist Plot")
plt.show()

sns.barplot(x='Day',y='Profit',data=df)
plt.title("Seaborn Bar Plot")
plt.show()

sns.scatterplot(x='Sales',y='Profit',data=df,color='yellow',edgecolor='black',size=30)
plt.title("Seaborn Scatter Plot")
plt.show()
```
